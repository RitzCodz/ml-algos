```python
import pandas as pd
import pylab as pl
from sklearn.cluster import KMeans

variables = pd.read_csv('cityoflondon.csv')
```

```python
df = variables[["CarsTaxis","PedalCycles"]]
import matplotlib.pyplot as plt
plt.scatter(df["CarsTaxis"], df["PedalCycles"])
plt.show()
```

```python
km = KMeans(n_clusters=3)
km.fit(df)
p = km.predict(df)
```

```python
pl.scatter(df["CarsTaxis"], df["PedalCycles"], c=km.labels_)

pl.xlabel('Cycles')
pl.ylabel('Cars and Taxis')
pl.title(' Cluster K-Means')
pl.show()
```
